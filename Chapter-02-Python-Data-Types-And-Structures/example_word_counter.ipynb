```python
from io import TextIOWrapper


def word_counter(fname: str) -> dict[str, int]:
    fhand: TextIOWrapper
    count: dict[str, int]
    words: list[str]

    try:
        fhand = open(fname, "r")
    except:
        print("File can not be opened")
        exit()

    count = dict()

    for line in fhand:
        words = line.split()
        for word in words:
            if word not in count:
                count[word] = 1
            else:
                count[word] += 1

    return count


# Testing
count: dict[str, int] = word_counter("alice.txt")
print("count:\n", count)
```

```python
# Using dictionary comprehension to construct the filtered dictionary
filtered: dict[str, int] = {
    key: value for (key, value) in count.items() if value < 20 and value > 18
}
print("Filtered:\n", filtered)
```
